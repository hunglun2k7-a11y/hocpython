```python
# Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) 
# lớn nhất nhưng nhỏ hơn n 
while True:
    try:
        n=int(input())
        if n>0:
            break
        else:
            print("vui long nhap so nguyen duong")
    except ValueError:
        print("vui long nhap so nguyen")
tong=0
b=1
while tong<n:
    tong+=b
    b+=1
print(b-2)
```

```python
#Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
# 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A
while True:
    try:
        A=int(input())
        if A>0:
            break
        else:
            print("vui long nhap so nguyen duong")
    except ValueError:
        print("vui long nhap so nguyen")
tong=0
b=1
while tong<A:
    tong+=1/b
    b+=1
print(b-2)
```

```python
#Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.
while True:
    try:
        n=int(input())
        if n>0:
            break
        else:
            print("vui long nhap so nguyen duong")
    except ValueError:
        print("vui long nhap so nguyen")
def songuyento(a):
    if a==1:
        return False
    elif a==2:
        return True
    for i in range(3,a):
        if a%i==0:
            return False
        return True
A=[]
while n>=1:
    phandu=n%10
    if songuyento(phandu):
        A.append(phandu)
    n=n//10
print(len(A))
```
