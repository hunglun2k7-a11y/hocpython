```python
#Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def songuyento(n):
    if n<2:
        return False
    elif n==2:
        return True
    for i in range(3,n):
        if n%i==0:
            return False
    return True
a=int(input())
if songuyento(a):
    print("day la so nguyen to")
else:
    print("day kh phai so nguyen to")
```

```python
#W5A4: VIết hàm kiểm tra một số có phải số hoàn hảo hay không
# xử lý trên dữ liệu kiểu số nguyên
#số hoàn hảo là số có các ước (trừ chính nó) có tổng bằng nó
def sohoanhao(n):
    a=[]
    if n==1:
        return True
    for i in range(1,(n//2)+1):
      if n%i==0:
            a.append(i)
    if sum(a)==n:
        return True
    return False
e=int(input())
if sohoanhao(e):
    print(e,"la so hoan hao")
else:
    print("day kh phai so hoan hao")
    
            
            
```

```python
#w5a9:Viết hàm trả về tổng các chữ số của số nguyên dương n
def tongcacso(n):
    a=[]
    while n>=1:
        phandu=n%10
        a.append(phandu)
        n=n//10
    return sum(a)
y=int(input())
print(tongcacso(y))
```

```python
#Viết hàm giả lập một máy tính bỏ túi.
#Hàm thực hiện tính toán kết quả của phép toán giữa số \
 #   ( num1 \) và \( num2 \) với toán tử\( operat \)
#Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
#Phép trừ \( - \)
#Phép cộng \( + \)
#Phép chia \( / \)
#Phép nhân \( * \)
#Kết quả làm tròn đến chữ số thập phân thứ hai.
#Đầu vào: Hai số thực và phép toán cần thực hiện
# (trong số các phép toán \(+, -, *, /\)).
#Đầu ra: Kết quả phép toán. 
a,b=map(int,input().split())
pheptoan=input()
def gialap(num1,num2,pheptinh):
    kq=None
    if pheptinh==("+"):
        kq=num1+num2
    elif pheptinh==("-"):
        kq=num1-num2
    elif pheptinh==("*"):
        kq=num1*num2
    elif pheptinh==("/"):
        kq=num1 / num2
    else:
        print("kh co phep tinh nhu vay")
    return(round(kq,2))
print(gialap(a,b,pheptoan))
```

```python
#w8:Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
def henhiphan(n):
    a=[]
    while n>0:
        phandu=n%2
        a.append(phandu)
        n=n//2
    a.reverse()
    return a       
x=int(input())
y=int(input())
list_a=henhiphan(x)
list_b=henhiphan(y)
c=0
if len(list_a)>len(list_b):
    for i in range(len(list_a)-len(list_b)):
        list_b.insert(0,0)
elif len(list_b)>len(list_a):
    for i in range(len(list_b)-len(list_a)):
        list_a.insert(0,0)
for i in range(len(list_a)):
    if list_a[i]!=list_b[i]:
        c+=1
print("khoang cach 2 so la" ,c)
#insert(vitri,giatri) them vao
         



    
```
