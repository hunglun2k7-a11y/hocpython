```python
# Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ
# thuật toán sắp xếp nào
def merge(a,b):
    c=[]
    j=0
    i=0
    while i<len(a) and j<len(b):
        if a[i] < b[j]:
            c.append(a[i])
            i+=1
        else:
            c.append(b[j])
            j+=1
    c.extend(a[i:])
    c.extend(b[j:])
    return c
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```

```python
#W9A5
def is_subarray(list1, list2):
    m, n = len(list1), len(list2)
    if m == 0:
        return True   
    if m > n:
        return False
    for i in range(n - m + 1):
        match = True
        for j in range(m):
            if list2[i + j] != list1[j]:
                match = False
                break
        if match:
            return True

    return False

print(is_subarray([2, 3, 4], [1, 2, 3, 4, 5]))
print(is_subarray([2, 4], [1, 2, 3, 4, 5])) 
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    i = 0
    while i < n:
        if flowerbed[i] == 0:
            left = (i == 0) or (flowerbed[i - 1] == 0)
            right = (i == n - 1) or (flowerbed[i + 1] == 0)
            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True
                i += 1
        i += 1
    return k <= 0
```

```python
#W9A8
def is_path_crossing(moves):
    x = 0
    y = 0
    seen = {(0, 0)}
    for ch in moves:
        if ch == 'R':
            x += 1
        elif ch == 'L':
            x -= 1
        elif ch == 'U':
            y += 1
        else:
            y -= 1
        if (x, y) in seen:
            return True
        seen.add((x, y))
    return False
```

```python
#W9A9
# [HARD - Tìm trung vị của 5 số]
# Viết hàm `def Median(nums):`
# Hàm nhận đầu vào là một danh sách (list) `nums` được đảm bảo
# có đúng 5 số nguyên.
# Yêu cầu:
# Tìm và trả về số trung vị (số lớn thứ ba) của 5 số này.
# Ràng buộc:
# 1. Chỉ được sử dụng tối đa 6 phép so sánh (ví dụ: <, >, <=, >=).
# 2. KHÔNG ĐƯỢC DÙNG bất kỳ hàm sắp xếp nào (như
# `sorted()`, `list.sort()`).
# Chú ý: Bạn chỉ cần viết hàm `Median` theo yêu cầu.
# Phần đọc input và gọi hàm (hàm main) đã có sẵn.
def Median(nums):
    a, b, c, d, e = nums

    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c

    if a > c:
        a, c = c, a
        b, d = d, b

    if b > e:
        b, e = e, b
    if b > c:
        b, c = c, b

    if b > d:
        b, d = d, b

    return b

```

```python
#W9A10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    j = 0
    res = 0
    m = len(heaters)

    for h in houses:
        while j + 1 < m and abs(heaters[j + 1] - h) <= abs(heaters[j] - h):
            j += 1
        d = heaters[j] - h
        if d < 0:
            d = -d
        if d > res:
            res = d
    return res

```
