```python
#Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần.
#Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):
def binary_search(arr,target):    
    left=0
    right=len(arr)-1
    while right>left:
        mid=(left+right)//2
        if target==arr[mid]:
            return mid
        else:
            if target < arr[mid]:
                right = mid - 1
            if target > arr[mid]:
                left = mid + 1
    return -1
arr=list(map(int,input().split()))
target=int(input())
print(binary_search(arr,target))
            
```

```python
#Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần trong danh sách. 
# def count_occurrences(arr, x) - > int:
def count_occurrences(arr,x):
    d=0
    for i in arr:
        if i==x:
            d+=1
    return d
arr=list(map(int,input().split()))
x=int(input())
print(count_occurrences(arr,x))
```

```python
#Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. 
# Sắp xếp nổi bọt (Bubble Sort);Sắp xếp chọn (Selection Sort)
# Thuật toán bubble sort,selection sort là thuật toán cho chạy for lồng so sánh lần lượt xem có cần phải swap hay kh?
# bubble sort , selection sort lần lượt là sắp xếp lớn trc bé sau và bé trước lớn sau
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0,len(arr)-1-i):
            if arr[j] > arr[j+1]:
                arr[j] , arr[j+1] = arr[j+1] ,arr[j]
    return arr
def selection_sort(arr):
    for i in range(len(arr)):
        minarr=i
        for j in range(i+1,len(arr)):
            if arr[minarr] > arr[j]:
                minarr=j
        arr[i] , arr[minarr] = arr[minarr] , arr[i]
    return arr
arr=list(map(int,input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))

                



```

```python
#Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. 
# Nếu có nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước.

n=list(map(int,input().split()))
dic={}
for i in n:
    if i not in dic:
        dic[i]=1
    else:
        dic[i]+=1
maxcounts=0
for i in dic:
    if dic[i] > maxcounts:
        maxcounts=dic[i]
for i in dic:
    if dic[i]==maxcounts:
        print(i,'xuat hien nhieu nhat , som nhat',maxcounts,', lan')
        break
    
```

```python
#Cho danh sách các số nguyên và một số X, đếm tất cả các cặp (a, b) sao cho a + b = X
n=list(map(int,input().split()))
X = int(input())
d=0
for i in range(len(n)):
    for j in range(i+1,len(n)):
        if n[i]+n[j]==X:
            d+=1
print(d)
```

```python
#Bạn được cung cấp một danh sách các danh sách lồng nhau, vd [[1, 2, 3], [4, 5], [6, 7, 8, 9]], hãy làm phẳng danh sách
import ast
n = input()
a=ast.literal_eval(n)
c=[]
for i in range(len(a)):
    if i%2!=0:
        a[i].reverse()
    for j in a[i]:
        c.append(j)
print(c)
```

```python
#Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất
# (không nhất thiết phải liền kề).
n=list(map(int,input().split()))
d=[]
for i in n:
    d.append(1)
for i in range(len(n)):
    for j in range(0,i):
        if n[i] > n[j]:
            d[i]=d[j]+1
print(d[i])
```

```python
#Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải nhỏ hơn nums[i].
#Trả về danh sách các số đếm.
n=list(map(int,input().split()))
c=[]
for i in range(len(n)):
    d=0
    for j in range(i+1,len(n)):
        if n[i] > n[j]:
            d+=1
    c.append(d)
print(c)
        
            
```
