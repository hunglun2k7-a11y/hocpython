```python
#W4A6
def prime(c):
    if c<2:
        return False
    elif c==2:
        return True
    else:
        for i in range(2,int(c**0.5)+1):
            if c%i==0:
                return False
    return True
a,b=map(int,input().split())
ans=0
for i in range(a,b+1):
    if prime(i):
        ans+=i
print(ans)
```

```python
#Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
#đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
#cách). Nếu không có, in ra (no number).
def doi1kn(a):
    b=[]
    while a>=1:
        phandu=a%10
        b.append(phandu)
        a=a//10
    for i in range(len(b)):
        for j in range(i+1,len(b)):
            if b[i]==b[j]:
                return False
    return True
def socp(a):
    if a<2:
        return False
    elif a==2:
        return True
    elif a%2==0:
        return False
    for i in range(3,a):
        if a%i==0:
            return False
    return True
n=int(input())
for i in range(n+1):
    if doi1kn(i) and socp(i):
        print(i,end=' ')

```

```python
import string
def string_piglatin(y):
    word=list(y)
    a={"u","e","o","a","i"}
    b=set(string.ascii_lowercase)-a
    if word[0] in a:
        word+=["w","a","y"]
        return "".join(word)
    elif word[0] in b:
        word.append(word[0])
        word.remove(word[0])
        word+=["a","y"]
        return "".join(word)
x=input()
print(string_piglatin(x))
```

```python
#Viết hàm giả lập máy tính bỏ túi thực hiện các phép cộng, trừ, nhân và chia 
#Hàm này trả về kết quả thực hiện toán tử. 
#Đầu vào: Ba biến cách nhau bởi dấu cách 
#Đầu ra: Giá trị sau khi thực hiện toán tử 
def calculate (num1, operator, num2): 
    kq=None
    if operator=="+":
        kq=num1+num2
    elif operator=="-":
        kq=num1-num2
    elif operator=="*":
        kq==num1*num2
    elif operator=="/":
        kq=num1/num2
    else:
        print("khong co phep tinh do")
    return round(kq,2)
x=int(input())
y=input()
z=int(input())
print(calculate(x,y,z))
```

```python
#Bài 13 
#Viết hàm tính ước số lẻ lớn nhất khác chính nó của một số nguyên dương n cho trước. 
#Đầu vào: Số nguyên n 
#Đầu ra: Ước số lẻ lớn nhất của n
def ulln(n):
    c=[]
    if n%2!=0:
        for i in range(1,(n//2)+1):
            if n%i==0:
                c.append(i)
        return(max(c))
a=int(input())
print(ulln(a))
```
