```python
#W11A1
def reverse(path: str) -> int:
    with open(path) as f:
        n = f.read().strip()
    return int(n[::-1])
```

```python
#w11A2
def canWinNim(path: str) -> bool:
    with open(path) as f:
        n = int(f.read().strip())
    return n % 4 != 0
```

```python
#A4
def areAnagrams(path1: str, path2: str) -> bool:
    def clean(s):
        return sorted(c.lower() for c in s if c.isalpha())

    with open(path1, encoding="utf-8") as f1, open(path2, encoding="utf-8") as f2:
        return clean(f1.read()) == clean(f2.read())
```

```python
#A8
def fibWords(n: int):
    f = ["a", "b"]
    for i in range(2, n+1):
        f.append(f[i-1] + f[i-2])
    with open("output.txt", "w") as out:
        for i in range(n+1):
            out.write(f[i] + "\n")
```

```python
#A9
def grade10(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # Bỏ dòng tiêu đề
    for line in lines[1:]:
        parts = line.strip().split(',')

        # Kiểm tra đủ số cột
        if len(parts) >= 7:
            ho = parts[1].strip()
            ten = parts[2].strip()
            thcs4 = parts[5].strip()

            if thcs4 == '10':
                print(ho + " " + ten)
```

```python
#A10
def getMoney(path: str) -> list[int]:
    with open(path) as f:
        return list(map(int, f.read().split()))

def rob(houses: list[int]) -> int:
    prev = curr = 0
    for x in houses:
        prev, curr = curr, max(curr, prev + x)
    return curr
```

```python
#A11
def findMinMax(path: str):
    try:
        with open(path) as f:
            nums = list(map(int, f.read().split()))
        print(min(nums), max(nums))
    except:
        print("Mission failed")
```

```python
#A12
def maximumProduct(path: str) -> int:
    with open(path) as f:
        f.readline()
        nums = sorted(map(int, f.read().split()))
    return max(nums[-1]*nums[-2]*nums[-3],
               nums[0]*nums[1]*nums[-1])
```

```python
#A13
def moveZeroes(path: str) -> list:
    with open(path) as f:
        f.readline()
        nums = list(map(int, f.read().split()))
    return [x for x in nums if x != 0] + [0]*nums.count(0)
```

```python
#A14
def averageTime(path: str):
    with open(path) as f:
        lines = f.readlines()[1:]
    times = [float(line.split(',')[1]) for line in lines]
    return round(sum(times)/len(times), 2)
```

```python
#A15
def findMovies(path: str) -> list[str]:
    res = []
    with open(path, encoding="utf-8") as f:
        lines = f.readlines()[1:]

    for line in lines:
        p = [x.strip() for x in line.split(',')]
        if float(p[-1]) > 8.0 and p[2].lower() != "boring":
            res.append(p[1])

    return res if res else ["All are boring!"]
```

```python
#A16
import math

def shapes(path: str):
    with open(path) as f:
        lines = [line.strip() for line in f if line.strip()]

    i = 0
    while i < len(lines):
        if "SQUARE" in lines[i]:
            side = float(lines[i+1].split(':')[1])
            print(f"{4*side:.2f}")
        elif "RECTANGLE" in lines[i]:
            parts = lines[i+1].split()
            w = float(parts[1])
            h = float(parts[3])
            print(f"{2*(w+h):.2f}")
        elif "CIRCLE" in lines[i]:
            r = float(lines[i+1].split(':')[1])
            print(f"{2*math.pi*r:.2f}")
        i += 2
```

```python
#A17
def productExceptSelf(path: str) -> list[int]:
    with open(path) as f:
        f.readline()
        nums = list(map(int, f.read().split()))

    n = len(nums)
    res = [1]*n
    left = right = 1

    for i in range(n):
        res[i] = left
        left *= nums[i]

    for i in range(n-1, -1, -1):
        res[i] *= right
        right *= nums[i]

    return res
```

```python
#A18
import math

def estimatedTime(path: str):
    with open(path) as f:
        lines = f.readlines()

    x1,y1 = map(float, lines[0].split()[2::2])
    x2,y2 = map(float, lines[1].split()[2::2])
    v = float(lines[2].split()[1])

    d = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return round(d/v, 2)
```

```python
#A19
import struct

def readBinary(path: str):
    with open(path, "rb") as f:
        while True:
            data = f.read(8)
            if not data:
                break
            print(f"{struct.unpack('d', data)[0]:.5f}")

```

```python
#A20
def imageSmoother(path: str):
    with open(path) as f:
        M = int(f.readline())
        img = [list(map(int, f.readline().split())) for _ in range(M)]

    res = [[0]*M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            s = c = 0
            for x in range(max(0,i-1), min(M,i+2)):
                for y in range(max(0,j-1), min(M,j+2)):
                    s += img[x][y]
                    c += 1
            res[i][j] = s//c

    for row in res:
        print(*row)
```
