```python
#W11A8
def fibonacci_words():
    try:
        n = int(input())
        if n < 0 or n > 20:
            return

        words = [""] * (n + 1)
        
        if n >= 0:
            words[0] = "a"
        if n >= 1:
            words[1] = "b"
            
        for i in range(2, n + 1):
            words[i] = words[i-1] + words[i-2]
            
        with open("output.txt", "w") as f:
            for word in words:
                print(word)
                f.write(word + "\n")
                
    except ValueError:
        pass

if __name__ == "__main__":
    fibonacci_words()
```

```python
#W11A10
def getMoney(path: str) -> list[int]:
    with open(path, 'r') as f:
        return [int(x) for x in f.read().split()]

def solve_robbery(money_list: list[int]) -> int:
    if not money_list:
        return 0
    if len(money_list) == 1:
        return money_list[0]
    
    prev1, prev2 = 0, 0
    for amount in money_list:
        temp = prev1
        prev1 = max(prev2 + amount, prev1)
        prev2 = temp
        
    return prev1
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    with open(path, 'r') as f:
        f.readline()
        nums = [int(x) for x in f.read().split()]
    
    nums.sort()
    return max(nums[-1] * nums[-2] * nums[-3], nums[0] * nums[1] * nums[-1])
```

```python
#W11A13
def moveZeroes(fileName: str) -> list:
    with open(fileName, 'r') as f:
        n = int(f.readline().strip())
        nums = [int(x) for x in f.readline().split()]
    
    result = [x for x in nums if x != 0]
    result.extend([0] * (len(nums) - len(result)))
    return result
    
```

```python
#W11A14
def averageTime(path: str) -> float:
    with open(path, 'r') as f:
        f.readline()
        total_time = 0
        count = 0
        for line in f:
            data = line.strip().split(',')
            total_time += float(data[1])
            count += 1
            
    if count == 0:
        return 0.0
        
    return round(total_time / count, 2)
```

```python
#W11A15
def findMovies(path: str) -> list[str]:
    result = []
    with open(path, 'r') as f:
        lines = f.readlines()

    for i in range(1, len(lines)):
        parts = lines[i].strip().split(',')
        if len(parts) >= 4:
            try:
                name = parts[1].strip()
                description = parts[2].strip()
                rating = float(parts[3].strip())
                
                if rating > 8.0 and description != "boring":
                    result.append(name)
            except ValueError:
                continue
                
    return result

if __name__ == "__main__":
    movies = findMovies("data.txt")
    print(movies)
```

```python
#W11A16
import math

def calculate_perimeter(path: str):
    with open(path, 'r') as f:
        lines = f.readlines()
    
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("TYPE:"):
            shape_type = line.split(":")[1].strip()
            i += 1
            if i < len(lines):
                data = lines[i].strip().split()
                p = 0.0
                
                if shape_type == "SQUARE":
                    side = float(data[1])
                    p = 4 * side
                elif shape_type == "RECTANGLE":
                    width = float(data[1])
                    height = float(data[3])
                    p = 2 * (width + height)
                elif shape_type == "CIRCLE":
                    radius = float(data[1])
                    p = 2 * math.pi * radius
                    
                print(f"{p:.2f}")
        i += 1

if __name__ == "__main__":
    calculate_perimeter("data.txt")
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r') as f:
        nums = [int(x) for x in f.read().split()]
    
    n = len(nums)
    result = [1] * n
    
    left = 1
    for i in range(n):
        result[i] = left
        left *= nums[i]
        
    right = 1
    for i in range(n - 1, -1, -1):
        result[i] *= right
        right *= nums[i]
        
    return result

if __name__ == "__main__":
    print(productExceptSelf("data.txt"))
```

```python
#W11A18
import math

def estimatedTime(path: str) -> float:
    with open(path, 'r') as f:
        lines = f.readlines()
        
    parts_src = lines[0].strip().split()
    x1 = float(parts_src[2])
    y1 = float(parts_src[4])
    
    parts_des = lines[1].strip().split()
    x2 = float(parts_des[2])
    y2 = float(parts_des[4])
    
    parts_vel = lines[2].strip().split()
    v = float(parts_vel[1])
    
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    time = distance / v
    
    return round(time, 2)

if __name__ == "__main__":
    print(estimatedTime("data.txt"))
```

```python
#W11A19
import struct
import os

def read_binary_file():
    path = input().strip()
    
    if os.path.exists(path):
        with open(path, 'rb') as f:
            while True:
                chunk = f.read(8)
                if len(chunk) < 8:
                    break
                number = struct.unpack('d', chunk)[0]
                print(f"{number:.5f}")

if __name__ == "__main__":
    read_binary_file()
```

```python
#W11A20
def imageSmoother(path: str):
    with open(path, 'r') as f:
        M = int(f.readline().strip())
        img = []
        for _ in range(M):
            img.append([int(x) for x in f.readline().split()])
            
    res = [[0] * M for _ in range(M)]
    
    for r in range(M):
        for c in range(M):
            total = 0
            count = 0
            for i in range(r - 1, r + 2):
                for j in range(c - 1, c + 2):
                    if 0 <= i < M and 0 <= j < M:
                        total += img[i][j]
                        count += 1
            res[r][c] = total // count
            
    for row in res:
        print(*row)

if __name__ == "__main__":
    imageSmoother("data.txt")
```
