```python
#Nhậ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma
#trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải
m=int(input())
n=int(input())
matrix=[]
for i in range(m):
    hangmoi=[]
    for j in range(n):
        gtri=int(input())
        hangmoi.append(gtri)
    matrix.append(hangmoi)
for i in matrix:
    for j in i:
        print(f'{j:>4}',end='     ')
    print()
    
```

```python
#Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
#đường chéo phụ của mảng
n=int(input())
matrix=[]
for i in range(n):
    hang=[]
    for j in range(n):
        a=int(input())
        hang.append(a)
    matrix.append(hang)
for i in matrix:
    for j in i:
        print(j,end="  ")
    print()
for i in range(n):
    for j in range(n):
        if i==j:
            print(f'{matrix[i][j]}',end="   ")
        elif i+j==n-1:
            print(f'{matrix[i][j]}',end='   ')
        else:
            print(f'{'    '}',end='')
    print()

        

```

```python
#Nhậ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên
#cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k)
m,n,k=map(int,input().split())
matrix=[]
for i in range(n):
    hang=[]
    for j in range(m):
        a=int(input())
        hang.append(a)
    matrix.append(hang)
for i in matrix:
    for j in i:
        print(f'{j:4}',end='')
    print()
b=0
for i in range(len(matrix)):
    b+=matrix[i][k-1]
print(b)



```

```python
#Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
#trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1
n=list(map(int,input().split()))
k=int(input())
for i in range(len(n)):
    if n[i]==k:
        print(i+1)
        break
print(-1)


```

```python
#Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
#chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện
n=list(map(str,input().split()))
dic={}
for i in range(len(n)):
    if n[i] not in dic:
        dic[n[i]]=[i]
    else:
        dic[n[i]].append(i)
for i in dic:
    dic[i]=tuple(dic[i])
print(dic)
```

```python
#Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
#trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9
n=list(map(int,input().split()))
a=[]
for i in n:
    if i==0:
        a.insert(0,0)
    else:
        a.append(i)
print(a)
        
```

```python
#Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
#nếu không có in ra -1
n=list(map(int,input().split()))
x=int(input())
c=[]
d=[]
for i in range(len(n)):
    if n[i]<x:
        c.append(i)
    elif n[i]>x:
        d.append(i)
if c==[]:
    print(-1)
else:
    for i in c:
        print(i)
```

```python
#Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
#thực sự, hay vô danh (không tăng, không giảm)
n=list(map(int,input().split()))
b=[]
for i in n:
    b.append(i)
n.sort()
c=[]
d=[]
for i in n:
    c.append(i)
n.reverse()
for i in n:
    d.append(i)
if c==d:
    print("vo danh")
elif b==c:
    print("day la danh sach tang")
elif b==d:
    print("day la danh sach giam")
else:
    print("vo danh")
```

```python
#Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
#phần tử đầu và cuối
n=list(map(int,input().split()))
c=[]
for i in range(1,len(n)-1):
    if n[i]>n[i-1] and n[i]>n[i+1]:
        c.append(n[i])
print(sum(c))
```
