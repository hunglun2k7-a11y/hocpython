```python
# Cho một mảng số nguyên không âm. Gọi
# mảng đó là "mảng đẹp" nếu mỗi phần tử
# x trong mảng xuất hiện đúng x lần.
# Kiểm tra xem mảng có phải là mảng đẹp
# hay không. Nếu không hãy xóa đi y số (y
# là nhỏ nhất có thể) để mảng còn lại thành
# mảng đẹp.
n=list(map(int,input().split()))
dic={}
n.sort()
for i in n:
    if i not in dic:
        dic[i]=[i]
    else:
        dic[i].append(i)
print(dic)
flag=False
d=0
for i in dic:
    a=len(dic[i])
    if i==a:
        flag=True
    else:
        if a>i:
            b=a-i
            d+=b
        else:
            d+=a
        flag=False

if flag==True:
    print("Yes")
else:
    print("no",d)

        


```

```python
# Sắp xếp có điều kiện: Số chẵn đứng
# trước, số lẻ đứng sau, chẵn tăng dần, lẻ
# giảm dần
n=list(map(int,input().split()))
c=[]
d=[]
for i in n:
    if i%2==0:
        c.append(i)
    else:
        d.append(i)
c.sort()
d.sort()
d.reverse()
c.extend(d)
for i in c:
    print(i,end=" ")
```

```python
# Tìm số có đúng k chữ số lớn nhất có thể
# trong một số có n chữ số (coi đầu vào là
# một số rất lớn - str). Không được hoán
# đổi vị trí các chữ số
n=list(input())
k=int(input())
vi_tri_bat_dau=0
d=[]
for i in range(k):
    vi_tri_ket_thuc=len(n)-k+i
    chu_so_max="0"
    vi_tri_do=0
    for j in range(vi_tri_bat_dau,vi_tri_ket_thuc+1):
        if int(chu_so_max)<int(n[j]):
            chu_so_max=n[j]
            vi_tri_do=j
    d.append(chu_so_max)
    vi_tri_bat_dau=vi_tri_do+1       
for i in d:
    print(i,end="")
```

```python
# đổi chỗ đúng 2 số để được số bé nhất
def doicho(n):
    a=list(n)
    b=-1
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if a[i] > a[j]:
                if i==0:
                    if b==-1 or a[j] =='0':
                        b=j
                        for k in range(j+1,len(a)):
                            if a[k]<=a[b]:
                                b=k
                else:
                    if b==-1 or a[b]>a[j]:
                        b=j
        if b != -1:
            a[i],a[b]=a[b],a[i]
            break
    b="".join(a)
    return int(b)
n=input()
print(doicho(n)) 
                
```

```python
def day(a,k):
    for i in range(len(a)-1,-1,-1):
        for j in range(i):
            if sum(a[j:i+1])/len(a[j:i+1])>k:
                return a[j:i+1]
    return "khong co day hop le"
a=list(map(int,input().split()))
k=int(input())
print(day(a,k))
```

```python
def so_nguyen_to(n):
    if n<2:
        return False
    elif n==2:
        return True
    elif n%2==0:
        return False
    else:
        for i in range(3,n):
            if n%i==0:
                return False
        return True
def tach_so(a):
    d=[]
    for i in range(2,a):
        while a%i==0 and so_nguyen_to(i):
            d.append(str(i))
            a=a//i
    return "*".join(d)
a=int(input())
print(tach_so(a))
        
            
```
