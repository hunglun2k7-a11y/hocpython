```python
# Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra kết quả
nums = list(map(int, input().split()))
result = []
seen = set()
for x in nums:
    if x not in seen:
        result.append(x)
        seen.add(x)

print(result)        


```

```python
# Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự
nums = list(map(int, input().split()))
result = []
s = 0
for x in nums:
    s += x
    result.append(s)

print(result)    

```

```python
# Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xá mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó
pairs = input().split()
result = {}
for p in pairs:
    key, value = p.split(":")
    if key not in result:
        result[key] = []
    result[key].append(value)
print(result)    
```

```python
# Nhập một list số nguyên, in ra dictionary{'positive':...,'negative':...,'zero':...} (đếm số >0,<0,=0)
nums = list(map(int, input().split()))
result = {
    'positives': sum(1 for x in nums if x > 0),
    'negatives': sum(1 for x in nums if x < 0),
    'zero': sum(1 for x in nums if x == 0)
}
print(result)
```

```python
# Viết chương trình nhập vào một tuple, in ra phần tử đầu, phần tử cuối và tuple đảo ngược
nums = tuple(map(int, input().split()))
print("Phan tu dau", nums[0])
print("Phan tu cuoi", nums[-1])
print("Tuple dao nguoc", nums[::-1])
```

```python
# Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary
strings = input().split()
result = {}
for s in strings:
    if s not in result:
        result[s] = 0
    result[s] += 1
print(result)    
```

```python
# Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key
dict1 = {}
for pair in input().split():
    k, v = pair.split(":")
    dict1[k] = int(v) 

dict2 = {}
for pair in input().split():
    k, v = pair.split(":")
    dict2[k] = int(v)

gop = dict1.copy()
for k, v in dict2.items():
    if k in gop:
        gop[k] += v
    else:
        gop[k] = v

sapxep_dict = {k: gop[k] for k in sorted(gop)}
print(sapxep_dict)


    


```
